```python
import pandas as pd
import numpy as np


# 1 ok
font = pd.read_excel('/content/drive/MyDrive/Acs/dados1_ac02.xlsx')
x = font.columns.tolist() # transforma cabeçalho em dados e em seguida em lista
y = font.values.tolist() # transforma dados em lista

x_serie = [cb for cb in x]

for aa in y:
  for ax in aa:x_serie.append(ax)

nova = [str(new)[:3] for new in x_serie]
x_serie = [float(new) for new in nova]


# 1.1 ok
new_series = pd.Series(x_serie)

# 1.2 ok
media = new_series.mean() # 5.9270000000000005
moda = new_series.mode().iloc[0] # 6.0
mediana = new_series.median() # 6.0

print(f'Média: {media}\nModa: {moda}\nMediana: {mediana}')

# 1.3 ok
variancia = new_series.var()
desvio = new_series.std()

Quartil_1 = new_series.quantile(0.25)
Quartil_2 = new_series.quantile(0.50)
Quartil_3 = new_series.quantile(0.75)

print(f'Variância: {variancia}\nDesvio: {desvio}\nQuartis: Q1={Quartil_1},  Q2={Quartil_2},  Q3={Quartil_1}')
```

```python
import pandas as pd
import numpy as np


# 2 ok
font = pd.read_excel('/content/drive/MyDrive/Acs/dados2_ac02.xlsx')

# 2.1 ok
alt = pd.DataFrame(font)
alt.columns = alt.iloc[7]
descarte = [0, 1, 2, 3, 4, 5, 6, 7]
alt = alt.drop(descarte)

colunas_ut = ['CO', 'O3', 'temp', 'umid']

# 2.1   2.2   2.3 ok
media_lt = ['media']
moda_lt = ['moda']
mediana_lt = ['mediana']
variancia_lt = ['variancia']
desvio_lt = ['desvio']
Quartil_01 = ['1°Quartil']
Quartil_02 = ['2°Quartil']
Quartil_03 = ['3°Quartil']
for cl in colunas_ut:
  media_lt.append(alt[cl].mean())
  moda_lt.append(list(alt[cl].mode()))
  mediana_lt.append(alt[cl].median())
  variancia_lt.append(alt[cl].var())
  desvio_lt.append(alt[cl].std())
  Quartil_01.append(alt[cl].quantile(0.25))
  Quartil_02.append(alt[cl].quantile(0.50))
  Quartil_03.append(alt[cl].quantile(0.75))


new_df = pd.DataFrame(columns=["---",'CO', 'O3', 'temp', 'umid'])
new_df.loc[0] = media_lt
new_df.loc[1] = moda_lt
new_df.loc[2] = mediana_lt
new_df.loc[3] = variancia_lt
new_df.loc[4] = Quartil_01
new_df.loc[5] = Quartil_02
new_df.loc[6] = Quartil_03


print(new_df)

new_df
```
